## O Pacote numpy
 Este pacote serve para operações matriciais e estatísticas básicas

```python
import numpy as np
```

```python
#criando números aleatórios

c = np.random.rand(100)
print(c)
print(type(c))

#acessando o último valor do array numpy
print(c[-1])

#acessando os 10 primeiros valores
print('os 10 primeiros valores da lista são: ' + str(c[0:10]))

for ii in range(0,10):
    print(str(ii) + ' '+ str(c[ii]))

#O comprimento do numpy array - shape da matriz
print(np.shape(c))
```

```python
mat3D = np.random.rand(365,100,100)

#Todos os valores da série
timeSeries = mat3D[:,49,49]
print(timeSeries)

# 10 últimos valores da série
timeSeries = mat3D[-10:,49,49]
print(timeSeries)

```

```python
#operações matriciais com numpy

timeSeries2 = mat3D[:,50:52,50]
print(np.shape(timeSeries2))
print(timeSeries2.shape)


timeSeriesTransposta = timeSeries2.transpose()
print(timeSeriesTransposta.shape)
```

```python
mediaTS = np.mean(timeSeries2)
print(mediaTS)

#
```

# Plotando com o matplotlib
- plot
- histogram
- bar
- boxplot
- pcolor

```python
#importando o pacote matplotlib
import matplotlib.pyplot as plt

#criando objeto matplotlib para plotar as figuras
fig,ax = plt.subplots(2)

#lineplot
ax[0].plot(mat3D[:,10,10])
ax[0].plot(mat3D[:,15:17,30])

# Bar
ax[1].bar(np.array(range(0,mat3D.shape[0])),mat3D[:,10,10]) 
```

```python
# boxplot

fig,ax = plt.subplots(2)
ax[0].boxplot(mat3D[:,10:20,0].flatten())
ax[1].boxplot(mat3D[:,10:20,0])
```

```python
#pcolor - gráfico de superfície

fig, ax = plt.subplots(2,2)

ax[0,0].pcolor(mat3D[0,:,:], cmap = 'jet')
ax[0,1].contour(np.sin(mat3D[0,:,:]))
ax[1,1].scatter(np.sin(mat3D[0,:,:]).flatten(),mat3D[0,:,:].flatten())
ax[1,0].hist(np.sin(mat3D[0,:,:]).flatten())
```
